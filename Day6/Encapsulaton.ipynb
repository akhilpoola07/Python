# Encapsulation

```python
class cls1():
    def __init__(self,a,b,c):
        self.a = a #public
        self._b = b #protected : to make the protected put the single underscore 
        self.__c = c#private : to make the protected put the double underscore 
obj1 = cls1(10,20,30)
print(obj1.a)#public can acess at anywhere using obj1.a
print(obj1._b)# if you know that the variable is protected then you can acess using same notation
#print(obj1.__c) can not be accesd using the same notation
try:
    print(obj1.__c)
except Exception:
    print("Can not be access in this way")
print(obj1._cls1__c) # private can be accessed using  this syntax called as Name mangling
```
