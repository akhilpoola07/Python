```python
file_representative = open("file1.txt",'r')
file1_data = file_representative.read()
print(file1_data)
```

Append Method

```python
file1_rep = open("file1.txt",'a')
file1_data = file1_rep.write("\nLine4")
file1_rep.close()
```

```python
with open("file1.txt","a") as e:
    e.write("\nLine7")
```

Write mode

```python
with open("file1.txt","w") as e:
    e.write("Line7")# remove the existing data and writing the new data
```
