1. Compile-time Polymorphism
Compile-time polymorphism means deciding which method or operation to run during compilation, usually through method or operator overloading.

Languages like Java or C++ support this. But Python doesn’t because it’s dynamically typed it resolves method calls at runtime, not during compilation. So, true method overloading isn’t supported, though similar behavior can be achieved using default or variable arguments.

Example: This code demonstrates method overloading using default and variable-length arguments. The multiply() method works with different numbers of inputs, mimicking compile-time polymorphism.

```python
# overloading
class Calculator:
    def multiply(self, a=1, b=1, *args):
        result = a * b
        for num in args:
            result *= num
        return result

# Create object
calc = Calculator()

# Using default arguments
print(calc.multiply())            
print(calc.multiply(4))           

# Using multiple arguments
print(calc.multiply(2, 3))       
print(calc.multiply(2, 3, 4))
```

```python
class socialmedia():
    def share_pic(self,platform):
        print(f"Sharing pic in {platform}")
    def post_pic(self,platform):
        print(f"Sharing post in {platform}")
class insta(socialmedia):
    pass
class facebook(socialmedia):
    pass
us1 = insta()
us2 = facebook()
us1.share_pic("instA")
us2.post_pic("fb")
us2.share_pic("fb")
```

# Runtime polymorphism 
means the method that gets executed is decided at runtime (during execution), not during coding.

This happens through method overriding — when a child class redefines a method of its parent class.

✅ Basic Concept
Parent class has a method
Child class overrides the same method
The method call depends on the object type, not the reference

```python
class Shape:
    def draw(self):
        print("Drawing shape")

class Circle(Shape):
    def draw(self):
        print("Drawing circle")

class Square(Shape):
    def draw(self):
        print("Drawing square")

def display(shape):
    shape.draw()

display(Circle())
display(Square())
```
