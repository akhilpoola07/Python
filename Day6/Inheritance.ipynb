# Single level Inheritrance

```python
class math():
    def add(self,a,b):
        return a+b
    def sub(self,a,b):
        return a-b,
class math2(math):
    def mul(self,a,b):
        return a*b
obj1 = math()
print(obj1.add(2,3))
obj2 = math2()
ans1 = obj2.add(2,3)# we are calling the fn present in the another class using object
print(ans1)
```

multiple inheritance

```python
class Father():
    def father_method(self):
        print("hello from Father")
class Mother():
    def mother_method(self):
        print("hello from mother method")
class child(Father,Mother):
    def child_method(self):
        print("Hello from child Method")
obj1 = child()
obj1.father_method()
obj1.mother_method()
obj1.child_method()
    
```

# Method Overridding

```python
class Father():
    def hello(self):
        print("From  fathermethod")
    def father_method(self):
        print("hello from Father")
class Mother():
    def hello(self):
        print("From mothermethod")
    def mother_method(self):
        print("hello from mother method")
class child(Father,Mother):#if we change the father,mother to mother ,father then the output will also change
    def child_method(self):
        print("Hello from child Method")
obj1 = child()
obj1.hello()
```

```python
class cs1():
    def __init__(self,a,b):
        self.a = a
        self.b = b
        print("a : ",a)
        print("b : ",b)
class cs2():
    def __init__(self,c,d):
        self.c = c
        self.d = d
        print("c : ",c)
        print("d : ",d)
class child(cs1,cs2):
    def __init__(self,a1,b1,c1,d1):#collect the values
        cs1.__init__(self,a1,b1)#distribute the values to class 1
        cs2.__init__(self,c1,d1)#distribute the values to class 2
        def hello():
            print("hello")
obj3 = child(1,2,3,4)
        
```

# Method resolution Order
| Method           | Behavior                                  |
| ---------------- | ----------------------------------------- |
| `cs2.__init__()` | Direct call (manual, no hierarchy)        |
| `super()`        | Follows **MRO (Method Resolution Order)** |
| Mixed approach   | Can break scalability in complex systems  |


```python
class cs1():
    def __init__(self, a, b, *args):
        self.a = a
        self.b = b
        print("a : ",a)
        print("b : ",b)
        super().__init__(*args)

class cs2():
    def __init__(self, c, d):
        print("cs2 init")
        self.c = c
        self.d = d
        print("c : ",c)
        print("d : ",d)

class child(cs1, cs2):
    def __init__(self, a1, b1, c1, d1):
        super().__init__(a1, b1, c1, d1)

obj = child(1, 2, 3, 4)
```

# Multilevel Inheritance


```python
# Base class
class Grandfather:
    def __init__(self, grandfathername):
        self.grandfathername = grandfathername

# Intermediate class
class Father(Grandfather):
    def __init__(self, fathername, grandfathername):
        self.fathername = fathername
        # Call the constructor of Grandfather
        Grandfather.__init__(self, grandfathername)

# Derived class
class Son(Father):
    def __init__(self, sonname, fathername, grandfathername):
        self.sonname = sonname
        # Call the constructor of Father
        Father.__init__(self, fathername, grandfathername)

    def print_name(self):
        print('Grandfather name :', self.grandfathername)
        print('Father name :', self.fathername)
        print('Son name :', self.sonname)

# Driver code
s1 = Son('Prince', 'Rampal', 'Lal mani')
print(s1.grandfathername)
s1.print_name()
```

# Hierarchial
one parent class and multiple chilld classes
When more than one derived class are created from a single base this type of inheritance is called hierarchical inheritance. In this program, we have a parent (base) class and two child (derived) classes.

```python
# Base class
class Parent:
    def func1(self):
        print("This function is in parent class.")

# Derived class 1
class Child1(Parent):
    def func2(self):
        print("This function is in child 1.")

# Derived class 2
class Child2(Parent):
    def func3(self):
        print("This function is in child 2.")

# Driver code
object1 = Child1()
object2 = Child2()

object1.func1()
object1.func2()
object2.func1()
object2.func3()
```

# hybrid
Hybrid inheritance is a combination of more than one type of inheritance. It uses a mix like single, multiple, or multilevel inheritance within the same program. Python's method resolution order (MRO) handles such cases.
Explanation:

School is the base class inherited by Student1, Student2, and Student3.
Student1 and Student2 each inherit School (single inheritance).
Student3 inherits both from Student1 and School (multiple inheritance).
Python handles method resolution using the MRO, so func1() from School is resolved without ambiguity.


```python
# Base class
class School:
    def func1(self):
        print("This function is in school.")

# Derived class 1 (Single Inheritance)
class Student1(School):
    def func2(self):
        print("This function is in student 1.")

# Derived class 2 (Another Single Inheritance)
class Student2(School):
    def func3(self):
        print("This function is in student 2.")

# Derived class 3 (Multiple Inheritance)
class Student3(Student1, School):
    def func4(self):
        print("This function is in student 3.")

# Driver code
obj = Student3()
obj.func1()
obj.func2()
```
