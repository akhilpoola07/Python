```python
class math():
    def add(self,a,b):
        return a+b
    def mul(self,a,b):
        return a*b
obj1 = math()# the object is an instance of class and Accessing the class
print("sum :",obj1.add(5,10))#use the object to access methods
obj2 = math()
print("Mul :",obj2.mul(2,3))
```

```python
class math():
    def __init__(self):#here you can pass the parameters
        #in python this is the constructor
        print("object created")
    def add(self,a,b):
        return a+b
    def mul(self,a,b):
        return a*b
obj1 = math()# the object is an instance of class and Accessing the class
#without calling any function the constructor method or fn automaticaay executed
#it executed whenever the object is called
```

```python
class math():
    def __init__(self):
        print("object created")
    def add(self,a,b):
        return a+b
    def mul(self,a,b):
        return a*b
obj1 = math()
obj2 = math()
print(obj1.add(1,3))
print(obj2.add(4,5))#you can change the arguments
```

```python

class Dog():
  
    def __init__(self,name,breed):
        self.a = name
        self.b = breed
        print(f"{self.a} is born")
    def bark(self):
        print(f"Name :{self.a} is born!")
        print("Woff!")
    def fetch(self,object):#here object is an argument
        print(f"got{object}")
my_dog1 = Dog("Max","pug")
my_dog1.bark()
print(my_dog1.a)
print(my_dog1.b)
my_dog1.a = "Tommy"
print(my_dog1.a)
my_dog1.fetch("stick")
```

```python
class MathDemo:
    def __init__(self,amount):
        self.amount=amount
        print(f"value : {amount}")
    def add_values(self,value):
        self.amount+= value
        print(f"After adding amount , value : {self.amount}")
    def remove_values(self,value):
        self.amount-=value
        print(f"After removing amount ,value : {self.amount}")
obj=MathDemo(1000)
obj.add_values(5000)
obj.remove_values(400)
```
