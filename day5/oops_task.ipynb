Problem: Bank Account Management System

Design a class BankAccount to simulate a simple banking system with the following features:

Initialize the account with:
Account holder name
Initial balance
Implement the following methods:
check_balance() → Display account holder name and current balance
deposit(amount) → Add money to the account
withdraw(amount) → Withdraw money with these conditions:
Cannot withdraw more than balance
Cannot withdraw negative or zero amount
The system should:
Prevent invalid transactions
Display updated balance after every operation

```python
class bank_account():
    def __init__(self,name,deposit):
        self.a = name
        self.b = deposit
    def Check_balance(self):
        print(f"Name : {self.a},balance : {self.b}")
    def Add_amount(self):
        c = int(input("enter the amount to add"))
        print(f"Added_Amount : {c}")
        self.b +=c
        print(f"Total balance : {self.b}")
    def withdraw_amount(self,withdraw):
        print(f"Withdraw_amount : {withdraw}")
        if self.b < withdraw:
            print("Insufficient balance")
        self.b -= withdraw
        print(f"Total balance : {self.b}")
usr1 = bank_account("Akhil",5000)
usr1.Check_balance()
usr1.Add_amount()
usr1.withdraw_amount(5000)


```
