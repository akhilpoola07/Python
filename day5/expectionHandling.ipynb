```python
#TO solve the above error by expecting using the excetion handling
# handle the runtime error
try:
    a = int(input("Enter first number"))
    b = int (input("enter second number"))
    c = a/b
    print(c)
except Exception:
    print("Error! Please check the Input once")
print("done!")
```
