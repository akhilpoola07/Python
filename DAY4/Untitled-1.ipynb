Loops

```python

for i in range(5):
    print(f"Line{i}")
```

```python
# loops in list
list1 = ["black Panther","natasha","Dr Strange","captain america","natasha"]
for i in list1:
    print(i)
# loops in string
print("out of the loop")
```

```python
#tuples
tp1 = (1,2,3)
for i in tp1:
    print("hello")
```

```python
top1 = ("a","b","c")
for i in top1:
    print("hello")
```

```python
str1 = "python"
for i in str1:
    print(i)    
```

for loop with dictionary

```python
dict1 = {"name":"akhil","age":24,"place":"hyderabad"}
for i in dict1:
    print("key :",i)#for keys
    print("values :",dict1[i])#for values
```

```python
for i in dict1.values():
    print(i)
```

```python
#List
list1 = [1,"Hello",3.14,True, 3+4j]
for i in list1:
    print(f"{i} is of type {type(i)}")
```

```python
transactions = [150.50,-30.00,50.75,-10.00,300.25]
sum = 0.00
for i in transactions:
    if (i > 0 ):
        sum += i
    else:
        transactions.pop(i)
print(f"Total sum of transactions {sum}")
```
