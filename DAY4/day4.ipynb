high order


```python
def fun1(a):
    return a *10
def fun2(fun,b):
    c = fun(b)
    return c
ans = fun2(fun1,6)
print(ans)
# explanation: fun1 is a function that takes a number and multiplies it by 10. fun2 is a higher-order function that takes another function (fun) and a value (b), calls the function with the value, and returns the result. In this case, fun2 calls fun1 with the value 6, which results in 60.
#fun = fun1
#fun2(fun,6)
```

```python
def wrapper_fun(fun):
    def modified_say_hi():
        fun()
        print("this is modified function")
    return modified_say_hi
def say_hi():
    print("Hello")
say_hi = wrapper_fun(say_hi)
say_hi()
```

```python
def wrapper_fun(fun):
    def modified_say_hi():
        fun()
        print("this is modified function")
    return modified_say_hi
@wrapper_fun #say_hi = wrapper_fun(say_hi)
def say_hi():
    print("Hello")
say_hi()
```

```python
def wrapper_fun(fun):
    def modified_fun(a,b):
        print("hello")
        fun(a,b)
        print("Done!")
    return modified_fun

@wrapper_fun
def add_fun(a,b):
    print(a+b) 

add_fun(10,20)
```
